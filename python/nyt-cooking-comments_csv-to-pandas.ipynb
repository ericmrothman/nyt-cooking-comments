```python
import os
import json
from pathlib import Path
import pandas as pd
import glob
import numpy as np

target_dir = Path.cwd().parent / 'csv' / '_processed'
export_dir = Path.cwd().parent / 'csv' / '_fromPandas'
csvs = glob.glob(os.path.join(target_dir, "*classified.csv"))

df_list = []
for csv in csvs:
    df_list.append(pd.read_csv(csv))
len(df_list)

df = pd.concat(df_list, ignore_index=True)
df
```

### [DATA CLEANING] Filter for approved authors

```python
authors_json = Path.cwd().parent / 'json' / '_for-scraping' / 'recipe_links_by_author.json'

with open(authors_json, 'r') as file:
    data = json.load(file)

authors = list(data.keys())

df = df[df['author'].isin(authors)]
df
```

### [DATA CLEANING] Separate rating from number of ratings

```python
df['recipeScore'] = df['recipeRating'].str.extract(r'([\d,.]+)[^(]*\(')
df['numberOfRatings'] = df['recipeRating'].str.extract(r'\(([\d,.]+)')
df = df.drop(columns='recipeRating')
df
```

### [DATA CLEANING] Adding sentiment column from map

```python
sentiment_map = {
    "positive_user_made_as_written" : "positive",
    "positive_user_describing_changes" : "positive",
    "thinks_recipe_is_flawed_due_to_flavor" : "negative",
    "thinks_recipe_is_flawed_due_to_technique_or_equipment" : "negative",
    "thinks_recipe_is_flawed_due_to_cooking_time" : "negative",
    "thinks_recipe_is_inauthentic" : "negative",
    "thinks_ingredients_hard_to_find" : "negative",
    "user_responding_to_other_user" : "neutral",
    "question" : "neutral",
    "maybe_not_about_the_recipe" : "neutral"
}

df['sentiment'] = df['ai_guess'].map(sentiment_map)
df
```

### [DATA CLEANING] Adding readable ai guess column

```python
readable_categories_map = {
    "positive_user_made_as_written" : "Liked recipe and prepared as written",
    "positive_user_describing_changes" : "Liked recipe but made changes",
    "thinks_recipe_is_flawed_due_to_flavor" : "Didn't like recipe due to flavor",
    "thinks_recipe_is_flawed_due_to_technique_or_equipment" : "Didn't like recipe due to technique or equipment",
    "thinks_recipe_is_flawed_due_to_cooking_time" : "Thought the cooking times were wrong",
    "thinks_recipe_is_inauthentic" : "Thought the recipe was inauthentic",
    "thinks_ingredients_hard_to_find" : "Thought ingredients were hard to find",
    "user_responding_to_other_user" : "Commenter responding to other commenter",
    "question" : "Commenter asking a question",
    "maybe_not_about_the_recipe" : "Off-topic"
}

df['readable_ai_guess'] = df['ai_guess'].map(readable_categories_map)
df
```

### [DATA CLEANING] Dropping prompt column

```python
df = df.drop(columns = 'prompt')
```

### (df_commenters) Commenters

```python
df_comment_recs = df.groupby('userID')['commentRecommendations'].sum()
df_comment_recs
```

```python
df_commenters = df.groupby('userID')['ai_guess'].value_counts().unstack().fillna(0)
df_commenters
```

```python
positive = ['positive_user_made_as_written', 'positive_user_describing_changes']
negative = ['thinks_recipe_is_flawed_due_to_flavor','thinks_recipe_is_flawed_due_to_technique_or_equipment','thinks_recipe_is_flawed_due_to_cooking_time','thinks_recipe_is_inauthentic','thinks_ingredients_hard_to_find']
neutral = ["user_responding_to_other_user","question","maybe_not_about_the_recipe"]

df_commenters = df_commenters.assign(
    Positive = lambda x: x[[col for col in positive if col in x.columns]].sum(axis=1),
    Negative = lambda x: x[[col for col in negative if col in x.columns]].sum(axis=1),
    Neutral  = lambda x: x[[col for col in neutral if col in x.columns]].sum(axis=1),
    posNegRatio = lambda x: x['Positive'] / x['Negative'].replace(0, np.nan)
)

df_commenters = df_commenters.merge(df_comment_recs, on='userID', how='right')
df_commenters = df_commenters.reset_index()
df_commenters
```

```python
len(df_commenters)
```

```python
all_comments = ['Positive','Negative','Neutral']
df_commenters['totalComments'] = df_commenters[all_comments].sum(axis=1)
df_commenters
```

```python
df_negative_commenters = df_commenters.query('posNegRatio < 1')
len(df_negative_commenters)
```

### (df_recipes) Comment categories per recipe

```python
url_map = df.set_index('recipeTitle')['url'].to_dict()

df_recipes = df.groupby('recipeTitle')['ai_guess'].value_counts().unstack().fillna(0)
total_comments = df_recipes.sum(axis=1)
df_recipes = df.groupby('recipeTitle')['ai_guess'].value_counts(normalize=True).unstack().fillna(0)
df_recipes['url'] = df_recipes.index.map(url_map)
df_recipes['totalComments'] = total_comments
df_recipe_sentiment = df.groupby('recipeTitle')['sentiment'].value_counts(normalize=True).unstack()
df_recipe_sentiment = df_recipe_sentiment.assign(posNegRatio=lambda df: df.positive/df.negative)
df_recipes = df_recipes.merge(df_recipe_sentiment, left_on='recipeTitle', right_index=True, how='left')
df_recipes = df_recipes.reset_index()
```

```python
df_recipes.to_csv(os.path.join(export_dir, "df_recipes.csv"), index=False)
```

```python
df_recipe_sentiment['recipeScore'] = df.groupby('recipeTitle')['recipeScore'].first()
df_recipe_sentiment['recipeScore'] = pd.to_numeric(df_recipe_sentiment['recipeScore'])
df_recipe_sentiment['recipeTitle'] = df.groupby('recipeTitle')['recipeTitle'].first()
df_recipe_sentiment['url'] = df.groupby('recipeTitle')['url'].first()
df_recipe_sentiment['recipeScore'].value_counts().reset_index()
df_recipe_sentiment
```
