```python
import os
import json
from pathlib import Path
import pandas as pd

target_dir = Path.cwd().parent



for root, dirs, files in os.walk(target_dir / 'json' / '_scraped'):
    for file in files:
        if file.endswith('.json'):
            with open(os.path.join(root, file), 'r') as f:
                data = json.load(f)
                rows=[]
                for index, recipe in enumerate(data):
                    stats = recipe.get('stats') or {}
                    rating = stats.get('Rating', "No rating found")
                    for comment in recipe.get('comments'): 
                        rows.append({
                            'index' : index,
                            'author' : recipe.get('author', 'No author found'),
                            'url' : recipe.get('url', 'No URL found'),
                            'recipeTitle' : recipe.get('title', 'No title found'),
                            'recipeRating' : rating,
                            'commentBody' : comment.get('commentBody', ''),
                            'commentType' : comment.get('commentType', ''),
                            'userID' : comment.get('userID', ''),
                            'commentRecommendations' : comment.get('recommendations', 0),
                        })
                df = pd.DataFrame(rows)
                df.to_csv(f'{file[12:-5]}.csv', index=False)
```
